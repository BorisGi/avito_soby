```python
%%time

links = []

for p in tqdm_notebook(range(1,101)):
    sleep(2)
    url = 'https://www.avito.ru/moskva_i_mo/sobaki?p='+str(p)+'&q=—Å–æ–±–∞–∫–∏'
        
    driver.get(url)
    text = str(driver.page_source)
    
    
    soup = BeautifulSoup(text, 'lxml')
    dogs = soup.findAll('div', class_='item__line')
    
    for i in dogs:
        link = 'https://www.avito.ru'+i.find('div', class_='snippet-title-row').find('a').get('href')
        links.append(link)
```

```python
# –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π –≤ —Ñ–∞–π–ª

with open('links.csv', 'w', encoding='UTF-8') as new_file:
    new_links = map(lambda x: x + '\n', links)
    new_file.writelines(new_links)
```

```python
# —Å–ø–∏—Å–æ–∫ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö

data = []
```

```python
%%time

for url in tqdm_notebook(links):
    sleep(3)
    driver.get(url)
    text = str(driver.page_source)
    soup = BeautifulSoup(text, 'lxml')
    try:
        topic = soup.find('div', class_='title-info-main').text.strip()
        breed = soup.find('div', class_='item-params').text.strip()
        price = soup.find('div', class_='item-price-wrapper').text.strip()
        adress = soup.find('div', class_='item-address').text.strip()
        text = soup.find('div', class_='item-description-text').text.strip()
        
        data.append([topic, breed, price, adress, text, url])
        
    except:
        print('–Ω–µ—Ç') 
```

```python
# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞–µ–º –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º

head = ['topic', 'breed', 'price', 'adress', 'text', 'url']
df = pd.DataFrame(data, columns=head)
```
